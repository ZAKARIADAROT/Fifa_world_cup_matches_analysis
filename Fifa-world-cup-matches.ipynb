```python
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'], errors='coerce')
```

```python
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'], format='%Y-%m-%d', errors='coerce')
```

```python
df_cleaned['total_goals'] = df_cleaned['number of goals team1'] + df_cleaned['number of goals team2']
goals_per_day = df_cleaned.groupby('date')['total_goals'].sum()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(goals_per_day.index, goals_per_day.values, marker='o')
plt.title("Total Goals Per Day")
plt.xlabel("Date")
plt.ylabel("Number of Goals")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Average possession for team1 and team2
avg_pos_team1 = df_cleaned['possession team1'].mean()
avg_pos_team2 = df_cleaned['possession team2'].mean()
```

```python
# Remove '%' and convert to float
df_cleaned['possession team1'] = df_cleaned['possession team1'].str.replace('%', '').astype(float)
df_cleaned['possession team2'] = df_cleaned['possession team2'].str.replace('%', '').astype(float)

# Now calculate the average
avg_pos_team1 = df_cleaned['possession team1'].mean()
avg_pos_team2 = df_cleaned['possession team2'].mean()

print("Average Possession Team 1:", avg_pos_team1)
print("Average Possession Team 2:", avg_pos_team2)
```

```python
plt.figure(figsize=(6, 5))
plt.bar(['Team 1', 'Team 2'], [avg_pos_team1, avg_pos_team2], color=['blue', 'green'])
plt.title("Average Possession Comparison")
plt.ylabel("Possession (%)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df_cleaned['total attempts team1'].dropna(), bins=20, color='orange', edgecolor='black')
plt.title("Distribution of Total Attempts (Team 1)")
plt.xlabel("Total Attempts")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_cleaned['total attempts team1'], df_cleaned['number of goals team1'], alpha=0.6, color='purple')
plt.title("Total Attempts vs Goals Scored (Team 1)")
plt.xlabel("Total Attempts")
plt.ylabel("Goals Scored")
plt.grid(True)
plt.tight_layout()
plt.show()
```
